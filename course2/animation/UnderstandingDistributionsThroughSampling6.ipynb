```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

%matplotlib notebook

fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2, 2, sharex=False, sharey=False, figsize=(8, 6))
# generate 4 random variables from the random, gamma, exponential, and uniform distributions

x1 = np.random.normal(0, 1, 1000)
x2 = np.random.gamma(2, 1.5, 1000)
x3 = np.random.exponential(2, 1000)
x4 = np.random.uniform(0 ,1, 1000)

def update(curr):
    # check if animation is at the last frame, and if so, stop the animation a
    if curr == 1000: 
        a.event_source.stop()

    ax1.cla()
    ax2.cla()
    ax3.cla()
    ax4.cla()
    
    ax1.axis([-3, 3, 0, 1])
    ax2.axis([-1, 12, 0, 1])
    ax3.axis([-1, 14, 0, 1])
    ax4.axis([-0.2 , 1.2, 0, 4])
    
    ax1.hist(x1[:curr], normed=True, bins=50, color='r', alpha=0.5)
    ax2.hist(x2[:curr], normed=True, bins=50, color='b', alpha=0.5)
    ax3.hist(x3[:curr], normed=True, bins=50, color='y', alpha=0.5)
    ax4.hist(x4[:curr], normed=True, bins=50, color='g', alpha=0.5)
    
    plt.text(-0.65, 9.5, 'Sampling: {}'.format(curr), fontsize=12)
    
    ax1.legend('Normal')
    ax2.legend('Gamma')
    ax3.legend('Exponential')
    ax4.legend('Uniform')

a = animation.FuncAnimation(fig, update, interval=1000, frames=range(100,1100,100))
```
